## Expressões Regulares

Expressões regulares são padrões usados para combinar ou encontrar ocorrências de sequências de caracteres em uma string. Em Python, expressões regulares são geralmente usadas para manipular strings e realizar tarefas como validação de entrada de dados, extração de informações de strings e substituição de texto.

```python
import re
```

```python
texto = "Meu e-mail é exemplo@gmail.com e você pode me contatar em outro_email@yahoo.com."
```

```python
# Expressão regular para contar quantas vezes o caracter arroba aparene no texto
resultado = len(re.findall("@", texto))
```

```python
print("O caractere '@' apareceru %d vezes no texto." %(resultado))
```

```python
# #xpressão regular para extrarir a palavra que aparece após a palavra "você" em um texto
resultado = re.findall(r'você (\w+)', texto)
```

```python
print("A palavra após 'você' é:", resultado[0])
```

> **Nota**: O r antes da string que representa a expressão regular em Python é usado para indicar que a string é uma string literal raw. Isso significa que as barras invertidas \ não são imterpretadas como caracteres de escape, mas são incluídas na expressão regular como árte do padrão

```python
# Expressão regular para extrair endereços de e-mail de uma string
emails = re.findall(r'\b[A-za-z0-9._%+-]+@[A-za-z0-9.-]+\.[A-z|a-z]{2,}\b', texto)
```

```python
print(emails)
```

[Documentação re()](https://docs.python.org/pt-br/3/library/re.html)

```python
text = "O aluno estava increvelmente perdido, mas encontrou a luz e rapidamente começou a aprender."
```

```python
# Extraindo os advérbios da fase
for m in re.finditer(r"\w+mente\b", text):
    print('%02d-%02d: %s' %(m.start(), m.end(), m.group(0)))
```

```python
# 1 - Crie um REGEX para contar quantas vezes o caracter "a" aparece em todo o texto da música.
qtd = len(re.findall(r'a', musica))
print("O caracter 'a' aparece %d no texto." %(qtd))
```

```python
# 2 - Crie um REGEX em Pytho para contar quantas vezes a palavra tempo aparece na música
qtd = len(re.findall(r'\btempo\b', musica, flags=re.IGNORECASE))
print("A palabra 'tempo' aparece %d vezes no texto." %(qtd))
```

```python
# 3 - Crie um REGEX em Python para extrair as palavras seguidas por exclamação.

palavras_exclamacao = re.findall(r'\b\w+!', musica)
print("Palavras seguidas por um ponto de exclamação:", palavras_exclamacao)
```

```python
# 4 - Crie um REGEX que extrai qualquer palavra cujo antecessor seja a palavra "esse" e o sucessor seja a palavra "amargo" em um texto.

palavras_correspondentes = re.findall(r'\besse\s+(\w+)\s+amargo\b', musica)
print("Palavra(s) encontradas:", list(palavras_correspondentes))
```

```python
# Usando re.findall() para encontrar todas as palavras com acento e apenas os caracteres antes do primeiro acento
palavras_com_acento = re.findall(r'\b[\wÀ-ÿ]+[áàâãéèêíïóôõöúüç]', musica)

print("Palavras com acento e caracteres antes do primeiro acento:", palavras_com_acento)
```
